```python
def prep_network_poi(nodes_df,edges_df):
    bad_pois = ['fast_food','restaurant','cafe','bar','bakery']
    good_pois = ['tree']
    pois = ['fast_food','restaurant','cafe','bar','bakery','tree']
#     pois = ['fast_food']
    place = 'Canton de Genve, Switzerland'
    edges_df_2056 = edges_df.to_crs('epsg:2056')
    edges_df_2056['geometry'] = edges_df_2056.buffer(5)
    for poi in pois:
        print(poi)
        if poi == 'bakery': 
            type_poi = 'shop'
        elif poi == 'tree':
            type_poi = 'natural'
        else:
            type_poi = 'amenity'

        tags = {type_poi:poi}
        print(tags)
        gdf = ox.geometries_from_place(place, tags)
        if 'element_type' in gdf.columns:
            gdf = gdf[gdf.element_type == 'node']
        gdf.index = gdf.index.map(int)
        gdf['id'] = gdf.index
        new_nodes, new_edges = toolbox.connect_poi(gdf, nodes_df, edges_df, key_col='id', path=None)
        snapped_pois = new_nodes[new_nodes.highway == 'projected_pap'].to_crs('epsg:2056')
        poi_by_street = gpd.sjoin(edges_df_2056,snapped_pois[['geometry']]).groupby('osmid_unique').size()
        edges_df[poi] = edges_df['osmid_unique'].map(poi_by_street).fillna(0)
        poi_imped = poi+'_'+'impedance'
        if poi in good_pois:
            edges_df[poi_imped] = edges_df['length'].astype(float) - edges_df[poi].astype(int)
        else:
            edges_df[poi_imped] = edges_df[poi].astype(int)*100 + edges_df['length'].astype(float)
    G = ox.graph_from_gdfs(nodes_df, edges_df)

    return G
G_poi = prep_network_poi(gdf_nodes,gdf_edges)
```
